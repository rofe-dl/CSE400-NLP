```python
import pandas as pd
import numpy as np
from speech_models import speech_random_forest, speech_naive_bayes, speech_logistic_regression
import ensemble

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report

from sklearn.feature_selection import chi2, SelectKBest
```

```python
df = pd.read_csv('file.csv')
df.head()
```

```python
features = df.drop(['emotion', 'file_name'], axis=1)
labels = df['emotion']

# features = features.filter(['mel_mean', 'mel_max', 'mel_std', 'rms_mean', 'rms_max', 'rms_std', 'spectral_flux_std', 'spectral_flux_mean', 'chroma_median', 'chroma_mean', 'mel_mean'], axis=1)

scaler = MinMaxScaler()
features[features.columns] = scaler.fit_transform(features[features.columns])

x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
# model = MLPClassifier(random_state=42, max_iter=5000)
# model = SVC(kernel='linear', probability=True, random_state=42)
model = RandomForestClassifier()
# model = speech_naive_bayes.get_naive_bayes()

# model = ensemble.StackEnsemble(meta_cls=speech_logistic_regression.get_logistic_regression())

model.fit(x_train, y_train)

results = model.predict(x_test)
print(classification_report(y_test, results))
```
